```python
import re

def extract_links_from_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        links = re.findall(r'(https?://\S+)', content)
        return links

# Exemplo de uso
file_path = '/content/playlist.txt'
links = extract_links_from_file(file_path)

# Exibir os links encontrados
for link in links:
    print(link)
```
